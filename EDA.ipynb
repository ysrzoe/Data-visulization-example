```python
import pandas as pd
df = pd.read_csv('C:/Users/ysr/Desktop/5218/num_df.csv')
df.head()
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

fig, axs = plt.subplots(1,3,figsize= (15,5))
age_denplot=sns.distplot(x=df['FICO'], ax= axs[0])
age_denplot.set_title("Distribution of FICO", fontsize=16)

exp_denplot=sns.distplot(x=df['Inquiries.in.the.Last.6.Months'], ax= axs[1])
exp_denplot.set_title("Distribution of Inquiries", fontsize=16)

inc_denplot=sns.distplot(x=df['Open.CREDIT.Lines'],ax=axs[2] )
inc_denplot.set_title("Distribution of parallel debts", fontsize=16)

plt.show()
```

```python
fig, axs = plt.subplots(1,3,figsize= (15,5))

job_denplot=sns.distplot(x=df['Monthly.Income'],ax=axs[0] )
job_denplot.set_title("Distribution of Monthly Income", fontsize=16)

hou_denplot=sns.distplot(x=df['Debt.To.Income.Ratio'],ax=axs[1] ) 
hou_denplot.set_title("Distribution of Debt To Income Ratio", fontsize=16)

age=sns.distplot(x=df['Amount.Requested'], ax= axs[2])
age.set_title("Distribution of Amount Requested", fontsize=16)

plt.show()
```

```python
fig, axs = plt.subplots(1,3,figsize= (15,5))

exp=sns.distplot(x=df['Amount.Funded.By.Investors'], ax= axs[0])
exp.set_title("Distribution of Amount Funded By Investors", fontsize=16)

inc=sns.distplot(x=df['Interest.Rate'],ax=axs[1] )
inc.set_title("Distribution of Interest Rate", fontsize=16)

hou=sns.distplot(x=df['Revolving.CREDIT.Balance'],ax=axs[2] ) 
hou.set_title("Distribution of Revolving CREDIT Balance", fontsize=16)

plt.show()
```
